### Задание 1

```python
import math

def select_center(s):
    c = round(len(s)/2) 
    if len(s)%2 ==0:
        res = s[c-1:c+1]
    else:
        res = s[c-1:c]
    return res

word = "test"       # Результат: es
print(select_center(word))
word = "testing"    # Результат: t
print(select_center(word))
```
